```python
# Estimate a standard work-hour assumption (e.g., 40 hours per job for fixed-price comparison)
ESTIMATED_HOURS = 40
```

```python
# compute average hourly salary where available

df['hourly_avg'] = df[['hourly_low','hourly_high']].mean(axis=1)
```

```python
import nltk
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Define stpopwords and common irrelevant words
custom_stopwords = set(stopwords.words('english'))
```

```python
print(df.columns)
```

```python
nltk.download('punkt_tab')
```

```python
# Function to clean and extract keywords from job titles

def extract_keywords(title):
  title = title.lower() # convert to lowercase
  title = re.sub(r'[^a-z\s]','',title) # remove special characters

  words = word_tokenize(title) # tokenize words
  words = [word for word in words if word not in custom_stopwords and len(word)>2] # Remove stopwords and short words
  return words

df['keywords'] = df['title'].apply(extract_keywords)
print("keywords column created")

# Now the rest of the code should work as expected
all_keywords = [word for sublist in df['keywords'] for word in sublist]
keyword_counts = Counter(all_keywords)

# Display the 20 most common keywords
print(keyword_counts.most_common(20))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extract keywords and frequencies for plotting
# Assign the 20 most common keywords to top_keywords
top_keywords = keyword_counts.most_common(20)

keywords, frequencies = zip(*top_keywords)

# Create a bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(keywords, frequencies)
plt.xlabel("Keywords")
plt.ylabel("Frequency")
plt.title("Top 20 Most Common Keywords in Job Titles")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Reuse the previous keyword extraction approach to categorize jobs

import re
from collections import Counter

# Function to clean and extract keywords from job titles
def extract_keywords_manual(title):
    title = title.lower()  # Convert to lowercase
    title = re.sub(r'[^a-z\s]', '', title)  # Remove special characters
    words = title.split()  # Split into words
    return words

# Apply function to extract keywords
df['keywords'] = df['title'].apply(extract_keywords_manual)

# Flatten keyword list and count occurrences
all_keywords = [word for sublist in df['keywords'] for word in sublist]
keyword_counts = Counter(all_keywords)

# Display the 20 most common keywords
keyword_counts.most_common(20)
```

```python
# Convert 'title' column to string and fill NaN values with an empty string
df['title'] = df['title'].astype(str).fillna('')

# Reapply keyword extraction
df['keywords'] = df['title'].apply(extract_keywords_manual)

# Flatten keyword list and count occurrences
all_keywords = [word for sublist in df['keywords'] for word in sublist]
keyword_counts = Counter(all_keywords)

# Display the 20 most common keywords
keyword_counts.most_common(20)
```

The extracted keywords include many common stopwords ("for," "and," "to"). I'll now remove these stopwords and reprocess the most frequent job-related keywords

```python
# Define a set of common stopwords to exclude
stopwords = set(["for", "and", "to", "a", "in", "with", "of", "the", "on", "at", "by", "is", "as", "an", "amp"])

# Filter out stopwords from keyword counts
filtered_keyword_counts = Counter({word: count for word, count in keyword_counts.items() if word not in stopwords})

# Display the top 20 job-related keywords after filtering
filtered_keyword_counts.most_common(20)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Convert and clean the date column
df.loc[:, 'published_date'] = pd.to_datetime(df['published_date'], errors='coerce')


# Extract Year-Month for trend analysis
df.loc[:, 'year_month'] = df['published_date'].dt.to_period('M')


# Count job postings per category over time
job_trend_data = df.groupby(["year_month", "title"]).size().unstack(fill_value=0)

# Convert 'year_month' index to datetime for time series modeling
job_trend_data.index = job_trend_data.index.to_timestamp()

# Identify the top 10 job roles with the highest total postings over time
top_roles = job_trend_data.sum().sort_values(ascending=False).head(10).index

# Filter the dataset to include only the top 10 job roles
top_roles_trend = job_trend_data[top_roles]
```

```python
top_roles_trend
```

```python
# Visualize the trends over time
plt.figure(figsize=(14, 6))
top_roles_trend.plot(marker='o', linestyle='-', figsize=(14, 6))
plt.title("Top 10 High-Demand Job Roles Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Job Postings")
plt.legend(title="Job Roles", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Select the most in-demand job role for forecasting
selected_role = top_roles[0]  # Choosing the top job role

# Prepare the time series data for modeling
job_series = top_roles_trend[selected_role]

# Split data into training and testing sets (80% train, 20% test)
train_size = int(len(job_series) * 0.8)
train, test = job_series[:train_size], job_series[train_size:]

# Train ARIMA model (Auto ARIMA can be used for optimal p, d, q selection)
model = ARIMA(train, order=(5, 1, 0))  # (p,d,q) parameters
model_fit = model.fit()
```

```python
# Forecast for the test period
forecast = model_fit.forecast(steps=len(test))
```

```python
# Evaluate model accuracy
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
```

```python
# Print model accuracy metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Plot actual vs predicted job postings
plt.figure(figsize=(12, 6))
plt.plot(job_series.index, job_series, label="Actual Job Postings", marker="o")
plt.plot(test.index, forecast, label="Predicted Job Postings", linestyle="dashed", marker="o", color="red")
plt.axvline(x=job_series.index[train_size], color="gray", linestyle="--", label="Train-Test Split")
plt.xlabel("Date")
plt.ylabel("Number of Job Postings")
plt.title(f"Forecasting Job Demand for {selected_role}")
plt.legend()
plt.grid(True)
plt.show()
```
